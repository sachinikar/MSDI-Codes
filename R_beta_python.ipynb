```python
#reading excel file
path='/content/drive/MyDrive/MSDI Index/4. Codes and Simulations/Revision2Codes/NC.xlsx'
```

```python
df = pd.read_excel(path)
require_cols = [0, 1]
required_df = pd.read_excel(path, usecols = require_cols)

require_col1 = [0]
Rainfall = pd.read_excel(path, usecols = require_col1)
Rainfall.columns=["Rainfall"]

#creating arrays
rainfall_array=Rainfall['Rainfall'].to_numpy()
```

```python
rescale_r=(rainfall_array-np.min(rainfall_array))/(np.max(rainfall_array)-np.min(rainfall_array))
```

```python
# finding the best fit distribution for rainfall #
dist_rainfall = distfit(distr=['beta'])
results_rainfall = dist_rainfall.fit_transform(rescale_r)

dist_rainfall.plot(cii_properties={'color': 'blue'})
```
